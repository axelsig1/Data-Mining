```python
#for low support, we have:
low_min_support = 0.001
sigma2 = low_min_support * num_records
confidence2 = 0.5
patterns2 = pyfpgrowth.find_frequent_patterns(associations, sigma2)
patterns3 = pyfpgrowth.find_frequent_patterns(associations, 10)
support2 = {key: value / len(df) for key, value in patterns2.items()}
rules2 = pyfpgrowth.generate_association_rules(patterns2, confidence2)

print(f'Nr of rules : {len(rules2)}')
print(f'nr of patterns: {len(patterns2)}')
print(f'nr of supports: {len(support2)}')
```

Calculate "lift":

$lift =  \frac{Confidence(A->B)}{Support(A)}$

Vi gr om skiten till en dataframe, dictionaries och tuples var sviidrygt att jobba med. 

```python
rows = []
for antecedent, (consequent, confidence) in rules2.items():
    antecedent_freq = patterns2.get(antecedent, 'NaN')   # support of antecedent 

    #Just a warnign thingy. 
    if antecedent_freq == 'NaN':
        print(f'the antecedent "{antecedent}" cant be found in patterns')
    else:
        antecedent_support = antecedent_freq / num_records
    
    #Since not all consequents will be found in the patterns, we need to calculate it manally.
    count_consequent = sum([1 for transaction in associations if all(item in transaction for item in consequent)])
    consequent_support = count_consequent/num_records
    #this can be done much better by creating a list of patterns with only 1 item in the patterns, but this shit works too.

    #print to check that it works:
    #print(f'Antedecent: {antecedent}, consequent: {consequent}, confidence: {confidence}, support of cosequent: {consequent_support}')
    
    
    # Calculate lift
    lift = confidence / consequent_support
    
    # Append to rows list
    rows.append({
        'Antecedent': antecedent,
        'Consequent': consequent,
        'Confidence': confidence,
        'Lift': lift
    })

# Create DataFrame
rules_df = pd.DataFrame(rows)
```

```python
# Lgger till columner som rankar efter lift och confidence
rules_df['Lift Rank'] = rules_df['Lift'].rank(method='first', ascending=False).astype(int)
rules_df['Confidence Rank'] = rules_df['Confidence'].rank(method='first', ascending=False).astype(int)
rules_df
```
